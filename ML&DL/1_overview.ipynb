## 머신러닝과 딥러닝

### 머신러닝 vs 딥러닝

- 머신러닝은 데이터의 특징을 스스로 추출하지 못하므로, 특성 추출 과정을 **인간이 직접 처리해주어야 함**

- 딥러닝은 **인간이 하던 작업을 생략**하고, 대량의 데이터를 신경망에 적용하여 컴퓨터가 스스로 분석한 후 답을 찾도록 함

- 머신러닝 안에 딥러닝이 포함되는 것이고, 이들은 AI를 구현하기 위한 방법임

### 머신러닝

- 컴퓨터가 스스로 대용량 데이터에서 특정 패턴을 찾아 학습하고, 그것을 토대로 새로운 데이터가 들어왔을 시 예측을 수행하는 것

#### 머신러닝 학습 과정

- 학습 데이터와 특정 알고리즘을 적용하여 학습시킴

- 학습 결과로 모델을 생성함

- 새로운 데이터를 모델에 적용하여 예측을 수행함

#### 머신러닝 학습 알고리즘

- 지도학습 : 정답을 알려주고 학습시키는 방법

  - 분류, 회귀
<br><br>
- 비지도학습 : 정답을 알려주지 않고 유사한 데이터끼리 범주화 하여 예측하는 방법

  - 군집화, 차원 축소

### 딥러닝

- 심층 신경망을 기반으로 하는 머신러닝의 일종

#### 딥러닝 학습 과정

- 데이터 준비

- 모델 정의

- 모델 컴파일 : optimizer, loss function 설정

- 모델 학습 : batch size, epoch 설정

- 예측

#### 딥러닝 학습 알고리즘

- 지도학습

  - 이미지 분류, 시계열 데이터 분석
<br><br>
- 비지도학습

  - 워드 임베딩, 군집화, 차원 축소
<br><br>
- 전이학습

  - pre-trained 모델을 기반으로 원하는 학습에 fine-tuning 하여 사용
